```python
import cv2
import cvzone
import mediapipe as mp
import os
from cvzone.HandTrackingModule import HandDetector


video = cv2.VideoCapture(0)     # Accessing the Webcam

video.set(3,640)
video.set(4,480)

hand_detection = HandDetector()  # creating instance for the class of HandDetector

image = cv2.imread(r"F:\CV_project  (tharun - 450)\Background image.png")    # reading the Background image


counter = 0
selection_speed = 10
item_selection = [(1136,196),(1136,555)]
count = 0
pause = 0


# Importing the photos of modes from the location 

A=[]
origin = r'F:\CV_project  (tharun - 450)\Modes'
for i in os.listdir(origin):
    modes = f'{origin}\{i}'
    A.append(cv2.imread(modes))



while video.isOpened():
    success,frame = video.read()
    
    if not success:
        break
    hands,frame = hand_detection.findHands(frame)     # reading the hands in the Frame
    image[172:480+172,136:136+640] = frame            # Adding the webcam access in the background image
    image[0:720,847:1280] = A[count]                  # Adding the modess in the backgorund image
    if hands and pause == 0:
        hand1 = hands[0]
        fingers = hand_detection.fingersUp(hand1)     # finding the fingers marks from the one hand
        print(fingers)

        # Veg or Non_veg
        
        if count == 0:
            # Non_Veg
            if fingers == [0,1,0,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,196),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 1
                    pause = 1


            # Veg
            elif fingers == [0,1,1,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,555),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 2
                    pause = 1
            else:
                counter = 0

                
#--------------------------------------------------------------------------------------------------------------------------------------------------

        # Egg or Fish
        
        elif count == 1:
            # Egg
            if fingers == [0,1,0,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,196),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 4
                    pause = 1


            # Fish
            elif fingers == [0,1,1,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,555),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 6
                    pause = 1

            else:
                counter = 0

                
#-----------------------------------------------------------------------------------------------------------------------------------------------                  
        
        # Peanut butter or Oats
        
        elif count == 2:
            # Bread_Peanut_Butter
            if fingers == [0,1,0,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,196),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 3
                    pause = 1


            # Oats
            elif fingers == [0,1,1,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,555),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 5
                    pause = 1
            else:
                counter = 0
#-------------------------------------------------------------------------------------------------------------------------------------------------

        # Selection of One or Two breads
        
        elif count == 3:
            # 1 bread
            if fingers == [0,1,0,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,196),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 7
                    pause = 1


            # 2 breads
            elif fingers == [0,1,1,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,555),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 7
                    pause = 1
            else:
                counter = 0

                
#------------------------------------------------------------------------------------------------------------------------------------------

        
         # Selection of One or Two Eggs
        
        elif count == 4:
            # 1 bread
            if fingers == [0,1,0,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,196),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 7
                    pause = 1


            # 2 breads
            elif fingers == [0,1,1,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,555),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 7
                    pause = 1


            else:
                counter = 0
#------------------------------------------------------------------------------------------------------------------------------------------
                
         # Selection of Oats with Fruits or Oats with Dry Fruits
        
        elif count == 5:
            # Oats with Fruits
            if fingers == [0,1,0,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,196),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 7
                    pause = 1


            # Oats with Dry Fruits
            elif fingers == [0,1,1,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,555),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 7
                    pause = 1

            else:
                counter

                
#------------------------------------------------------------------------------------------------------------------------------------------

# Selection of one fish meat or 2 fish meats
        
        elif count == 6:
            # One fish meat
            if fingers == [0,1,0,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,196),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 7
                    pause = 1


            # Two fish meats
            elif fingers == [0,1,1,0,0]:
                counter += 1
                cv2.ellipse(image,(1136,555),(125,125),0,0,counter*selection_speed,(0,255,0),10)

                if counter * selection_speed >= 360:
                    counter = 0
                    count = 7
                    pause = 1
            else:
                counter = 0

                
            
    if pause > 0:
        pause+=1    
        if pause > 90:
            pause = 0
            
            
    
    cv2.imshow('1',image)
    
    if cv2.waitKey(1) == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
